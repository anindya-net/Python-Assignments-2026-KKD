## Random Numbers:

### 1. Generate 100 integer random numbers that follow uniform distribution between (1,100). Then, find the mean

```python
import numpy as np
import random
np.random.seed(42)
distribution=np.random.randint(1,101, size=100)
print("Distribution:",distribution)
print("Mean:",np.mean(distribution))
```

### 2. Find 100 random number that follow Gaussian distribution (normal distribution) with mean “zero” and standard deviation “1”. Then calculate the mean and standard deviation of those numbers.

```python
np.random.seed(50)
mu=0. #mean
sigma=1. #standard deviation
gaussian=np.random.normal(mu,sigma,size=100)
print("Distribution: ",gaussian)
print("Mean:", np.mean(gaussian))
print("Standard deviation:", np.std(gaussian))
```

### 3. Simulate a battle between two players rolling a 6-sided die. The player with the higher roll wins the round. Determine the winner after 10 rounds.

```python
val1=np.random.randint(1,7)
val2=np.random.randint(1,7)
print("Player1: ",val1)
print("Player2: ",val2)
if val1>val2:
    print("Player 1 wins")
elif val1==val2:
    print("It's a draw")
else:
    print("Player2 wins")
print("After 10 Rounds: ")
for i in range (1,11,1):
    val1=np.random.randint(1,7)
    val2=np.random.randint(1,7)
print("Player1: ",val1)
print("Player2: ",val2)
if val1>val2:
    print("Player 1 wins")
elif val1==val2:
    print("It's a draw")
else:
    print("Player2 wins")
```

## Plotting with Matplotlib

### 5. Write a Python program to plot the function: f(x)= sin(x)*exp(-x^2) for x values in the range [-3, 3]. Label the axes and add a title. Make the linewidth 2, linestyle ‘--‘, blue color. Save the plot in png format.

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(-3,3,1000)
f=np.sin(x)*np.exp(-x**2)
plt.figure(figsize=(16,9))
plt.plot(x,f,label= "Function plot",linewidth=2,color='blue',linestyle="--")
plt.xlabel("X")
plt.ylabel("f(x)")
plt.title("Gaussian Attenuated Sine Wave")
plt.savefig("plot1.png")
plt.show()
```

### 6. Plot the following functions on the same graph for x in the range [-5, 5], give legends,title, x,y labels and save it in a single plot. 
### f(x)=x^2 (blue, dashed line)
### g(x)=x^3 (red, solid line)
### h(x)=exp(-x) (green, dotted line)

```python
import matplotlib.pyplot as plt
import numpy as np
x= np.linspace(-5,5,100)
f=x**2
g=x**3
h=np.exp(-x)
plt.figure(figsize=(16,9))
plt.plot(x,f,label= "f(x) ",linewidth=2,color='blue',linestyle="--")
plt.plot(x,g,label= "g(x) ",linewidth=2,color='red',linestyle="-")
plt.plot(x,h,label= "h(x) ",linewidth=2,color='green',linestyle=":")

plt.xlabel("X")
plt.ylabel("Function")
plt.title("function vs variable")
plt.legend()
plt.savefig("plot2.png")
plt.show()
```

### 8. (i) Generate a dataset that contains 1000 random Gaussian numbers with mean zero and standard deviation “1”, store them in an array. Then, write a Python program to plot a histogram with 10 bins and save the plot.
